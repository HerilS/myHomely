0th Tab

```python
df = pd.read_excel('Data Dictionary SuperDemographics 2020.xlsx',sheet_name = 'Population', header = 1)
```

```python
df['length'] = df['Description'].apply(lambda x: len(x))
```

```python
df['Variable'] = df['Variable'].apply(lambda x: x.strip())
df['Description'] = df['Description'].apply(lambda x: x.strip())
df['Description'] = df['Description'].apply(lambda x: 'Population' + ' ' + x if 'population' not in x else x)
```

```python
df['Description-1stpart'] = df['Description'].apply(lambda x: x.split(':')[0])
```

```python
df
```

```python
for value in df[df.length < 100].Variable.values:
    datadictionary[value] = df[df.Variable == value]['Description'].item()
for value in df[df.length > 100].Variable.values:
    datadictionary[value] = df[df.Variable == value]['Description-1stpart'].item()
```

```python
datadictionary
```

```python
for feat in df[df.length > 100].Variable.tolist():
    print(feat)
    print(df[df['Variable'] == feat]['Description'].item())
    print('')
```

```python
for feat in df[df.length > 100].Variable.tolist():
    print(feat)
    print(df[df['Variable'] == feat]['Description-1stpart'].item())
    print('')
```

```python
df = df.iloc[1:,:]
```

```python
df['length'] = df['Description'].apply(lambda x: len(x))
```

```python
df['Variable'] = df['Variable'].apply(lambda x: x.strip())
df['Description'] = df['Description'].apply(lambda x: x.strip())
```

```python
def temp(x,y):
    if '15' in y:
        return 'population aged 15 years and over: ' + x
    elif '15' in y:
        return 'population aged 15 - 24 years: ' + x
    elif '15' in y:
        return 'population aged 25 - 64 years: ' + x
    else:
        return x
```

```python
df['Description'] = df.apply(lambda x: temp(x.Description, x.Variable), axis=1)
```

```python
df['Description-1stpart'] = df['Description'].apply(lambda x: x.split(':')[0])
```

```python
for value in df[df.length < 100].Variable.values:
    datadictionary[value] = df[df.Variable == value]['Description'].item()
for value in df[df.length > 100].Variable.values:
    datadictionary[value] = df[df.Variable == value]['Description-1stpart'].item()
```
